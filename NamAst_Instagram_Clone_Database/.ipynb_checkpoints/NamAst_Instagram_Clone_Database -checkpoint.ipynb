```python
#3 We want to target our inactive users with an email campaign.
#  Find the users who have never posted a photo
%sql SELECT  username FROM users u   WHERE u.id not in (SELECT  user_id FROM photos)
```

```python
%sql SELECT username FROM users LEFT JOIN photos ON users.id = photos.user_id WHERE photos.id IS NULL;
```

```python
# 4. We're running a new contest to see who can get the most likes on a single photo.
#WHO WON??!!
%sql SELECT username,photos.id,photos.image_url,COUNT(*) AS Total_Likes  FROM likes JOIN photos  ON likes.photo_id = photos.id JOIN users ON users.id = photos.user_id  GROUP BY photos.id  ORDER BY Total_Likes DESC LIMIT 1   
# version 2
```

```python

# Our Investors want to know...

#5. How many times does the average user post?*/
# total number of photos/total number of users
%sql SELECT ROUND((SELECT COUNT(*)FROM photos)/(SELECT COUNT(*) FROM users),2);

```

```python
#6. user ranking by postings higher to lower
%sql SELECT users.username,COUNT(photos.image_url) FROM users JOIN photos ON users.id = photos.user_id GROUP BY users.id ORDER BY 2 DESC;
```

```python
#7 .Total Posts by users (longer versionof SELECT COUNT(*)FROM photos) 
%sql SELECT SUM(total_posts_per_user.TotalPost) FROM  (SELECT u.id , u.username, COUNT(*) AS TotalPost FROM users u JOIN photos p ON p.user_id = u.id GROUP BY u.id ) AS total_posts_per_user
```

```python
#8 /*total numbers of users who have posted at least one time */
%sql SELECT  COUNT(DISTINCT(u.id)) FROM users u JOIN photos p ON p.user_id = u.id 
```

```python
#9 A brand wants to know which hashtags to use in a post
# What are the top 5 most commonly used hashtags?
%sql SELECT tags.tag_name , COUNT(*) AS Total FROM tags  INNER JOIN photo_tags ON photo_tags.tag_id = tags.id GROUP BY tags.id ORDER BY Total DESC LIMIT 5
```

```python
#10 We have a small problem with bots on our site...Find users who have liked every single photo on the site
%sql SELECT u.id,u.username  , COUNT(*) AS total FROM users u INNER JOIN likes ON u.id = likes.user_id GROUP BY u.id HAVING total = (SELECT COUNT(*) FROM photos )
```

```python
#11 We also have a problem with celebrities . Find users who have never commented on a photo

%sql SELECT u.username , u.id , comments.comment_text FROM users u  LEFT JOIN comments ON u.id = comments.user_id GROUP BY u.id HAVING comment_text IS NULL;
```

```python
#12 Mega Challenges
# Are we overrun with bots and celebrity accounts?
# Find the percentage of our users who have either never commented on a photo or have commented on every photo*/
%sql  SELECT tableA.total_A AS 'Number Of Users who never commented',   (tableA.total_A/(SELECT COUNT(*) FROM users))*100 AS '%',        tableB.total_B AS 'Number of Users who likes every photos',(tableB.total_B/(SELECT COUNT(*) FROM users))*100 AS '%'FROM(SELECT COUNT(*) AS total_A FROM(SELECT username,comment_text FROM users LEFT JOIN comments ON users.id = comments.user_id GROUP BY users.id HAVING comment_text IS NULL) AS total_number_of_users_without_comments) AS tableA JOIN(   SELECT COUNT(*) AS total_B FROM(   SELECT TABLEC.id , COUNT(TABLEC.photo_id) AS total FROM (	SELECT DISTINCT users.id,comments.photo_id FROM users LEFT JOIN comments ON comments.user_id = users.id  ) AS TABLEC GROUP BY TABLEC.id HAVING total = (SELECT COUNT(*) FROM photos  )) AS users_comments_every_photos) AS TABLEB


```
